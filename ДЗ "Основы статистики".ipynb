**Задание**

В этом задании нужно убедиться, что ЦПТ действительно работает.

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n. Постройте гистрограммы средних этих выборок и сделайте выводы.

**Решение**

```python
import pandas as pd #Загрузила библиотеки
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#Выбрала экспоненциальное распределение с параметрами 15 и 1.3

rv = sts.expon(loc=15,scale=1.3)
sample = rv.rvs(size=1000) #Сгенерировала 1 000 выборок с помощью метода rvs
```

```python
#Построила гистограмму

x = np.linspace(0,2,100)
pdf = rv.pdf(x)

plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, f(x)')
plt.xlabel('x')
plt.show()
```

```python
n1 = 2                #Размеры выборок
n2 = 20
n3 = 99

a1 = np.array([])     #Массивы для сохранения средних по выборкам
a2 = np.array([])
a3 = np.array([])

samples_count = 1000  #Количество повторов генерации

for number in range(samples_count):
    sample1 = rv.rvs(size=n1)               #На каждом цикле сгенерировала выборки указанных размеров
    sample2 = rv.rvs(size=n2)
    sample3 = rv.rvs(size=n3)
    a1 = np.append(a1, sample1.mean())      # Посчитала среднее по каждой выборке и добавила в массив средних
    a2 = np.append(a2, sample2.mean())
    a3 = np.append(a3, sample3.mean())
```

```python
label_1 = 'sample of ' + str(n1) + ' histogram'   #Подписи для графиков
label_2 = 'sample of ' + str(n2) + ' histogram'
label_3 = 'sample of ' + str(n3) + ' histogram'

plt.hist(a1, density=True, label=label_1)         #Построила графики средних выборок n1, n2 и n3
plt.hist(a2, density=True, label=label_2)
plt.hist(a3, density=True, label=label_3)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('x')
```

Итог. Мы видим на графике, что распределение случайных величин выборок n1, n2 и n3 стало близко к нормальному.
